```python
newslist=[]
newswordlist=[]
feed='https://www.google.com/alerts/feeds/17765963020151138991/3534812369723724452'
f=feedparser.parse(feed)
for e in f.entries:
  print('\n---------------------------')
  text=e.title+"\n"+e.description
  print(text)
  endMarkup=e.description.find('</a>')                # Finde Ende vom Markup
  rawDescription=e.description[endMarkup+4:]
  text=e.title+"\n"+rawDescription           #Zusammenführen von Titel und Beschreibung
  newslist.append(text)  
  print(text)
  textlist=[w.lower().strip('.,!?.,') for w in text.split()]
  print(textlist)
  newswordlist.append(textlist)
print('Number of news in feed:',len(newslist))
```
